```python
x=[]
y=[]
n=int(input("Enter number of inputs"))
for _ in range(n):
    print("Enter x{}value".format(_))
    x.append(int(input()))
    print("Enter y{}value".format(_))
    y.append(int(input()))
```

```python
import matplotlib.pyplot as plt 
plt.xlabel('x') 
plt.ylabel('y')
plt.plot(x, y,'r',)
plt.grid() 
plt.show()
```
